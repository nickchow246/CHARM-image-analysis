```python
# use cellpose model to predict 'crop' file
# export file
from cellpose import io, models

file = r"C:\Users\bear\Desktop\liver segmentation\crop_nucleus.tif"
model_yx_nucleus_path = r"C:\Users\bear\Desktop\liver segmentation\models\yx_nucleus_model"
model_ortho_nucleus_path = r"C:\Users\bear\Desktop\liver segmentation\models\ortho_nucleus_model"

io.logger_setup()
model = models.CellposeModel(gpu = True, pretrained_model = model_yx_nucleus_path, pretrained_model_ortho = model_ortho_nucleus_path)
image_cellpose = io.imread(file)
print(image_cellpose.shape)

channels = [0, 0]
masks, flows, styles = model.eval(image_cellpose, channels = channels, do_3D = True)
io.save_masks(image_cellpose, masks, flows, file, tif=True, channels = channels, suffix='_trained_3d')
```

```python
# convert one channel per file to multichannel tiff
import os
import numpy as np
import tifffile

# please set manually
folder = r"C:\Users\bear\Desktop\new 3D HnE\bright field, deglitch (6 organs)"
base = "tongue muscle resized" # exclude '_c[i].tif'
ext = ".tif"
num_channel = 3

channels = []
for i in range(num_channel):
    filepath = os.path.join(folder, f"{base}_c{i}{ext}")
    img = tifffile.imread(filepath)
    channels.append(img)

combined = np.stack(channels, axis=1)
z, c, y, x = combined.shape
assert c == min(z, c, y, x)

dst = os.path.join(folder, f"{base}_multichannel{ext}")
tifffile.imwrite(dst, combined, imagej=True, metadata={'axes': 'ZCYX'})
```

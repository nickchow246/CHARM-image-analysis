```python
import numpy as np
from skimage.morphology import erosion, dilation, ball
import tifffile

img_cell_path = r"C:\Users\bear\Desktop\figure 4\crop_250.tif"
img_nucleus_path = r"C:\Users\bear\Desktop\figure 4\crop_250_nucleus.tif"
cell_mask_path = r"C:\Users\bear\Desktop\figure 4\crop_250_temp.tif"
nucleus_mask_path = r"C:\Users\bear\Desktop\figure 4\crop_250_nucleus_temp.tif"

img_cell = tifffile.imread(img_cell_path)
cell_mask = tifffile.imread(cell_mask_path)
nucleus_mask = tifffile.imread(nucleus_mask_path)
print(img_cell.shape)
print(cell_mask.shape)
print(nucleus_mask.shape)
```

```python
def open_then_close(img, radius):
    footprint = ball(radius)
    img = erosion(img, footprint = footprint)
    img = dilation(img, footprint = footprint)
    img = dilation(img, footprint = footprint)
    img = erosion(img, footprint = footprint)
    return img

cell_open = open_then_close(cell_mask, radius = 2)
nucleus_open = open_then_close(nucleus_mask, radius = 1)
```

```python
print(img_cell.shape)
z, c, y, x = img_cell.shape
composite = np.zeros((z, 4, y, x))
composite[:, 0, :, :] = img_cell[:, 0, :, :] # nucleus
composite[:, 1, :, :] = img_cell[:, 1, :, :] # cell
composite[:, 2, :, :] = cell_open
composite[:, 3, :, :] = nucleus_open
print(composite.shape)
print(composite.dtype)

dst = r"C:\Users\bear\Desktop\figure 4\composite_python_temp.tif"
tifffile.imwrite(dst, composite.astype(np.float32), imagej=True, metadata = {'axes': 'ZCYX'})
```

```python
# convert multichannel tiff to one channel per file
import tifffile
import os

multichannel_path = r"C:\Users\bear\Desktop\figure 4\matlab_mode.tif"

folder, filename = os.path.split(multichannel_path)
base, ext = os.path.splitext(filename)
multichannel = tifffile.imread(multichannel_path)
z, c, y, x = multichannel.shape
assert c == min(z, c, y, x)

for i in range(c):
    channel = multichannel[:, i, :, :]
    dst = os.path.join(folder, f"{base}_c{i}{ext}")
    tifffile.imwrite(dst, channel, imagej = True, metadata = {'axes': 'ZYX'})
```

```python
# convert one channel per file to multichannel tiff
import os
import numpy as np
import tifffile

# please set manually
folder = r"C:\Users\bear\Desktop\figure 4"
base = "highlight_polyploidy" # exclude '_c[i].tif'
ext = ".tif"
num_channel = 7

channels = []
for i in range(num_channel):
    filepath = os.path.join(folder, f"{base}_c{i}{ext}")
    img = tifffile.imread(filepath)
    channels.append(img)

combined = np.stack(channels, axis=1)
z, c, y, x = combined.shape
assert c == min(z, c, y, x)

dst = os.path.join(folder, f"{base}_multichannel{ext}")
tifffile.imwrite(dst, combined, imagej=True, metadata={'axes': 'ZCYX'})
```

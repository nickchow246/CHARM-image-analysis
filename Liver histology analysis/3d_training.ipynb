```python
from cellpose import io, models

file = r"C:\Users\bear\Desktop\liver segmentation\crop.tif"
model_yx_path = r"C:\Users\bear\Desktop\liver segmentation\models\yx_model"
model_ortho_path = r"C:\Users\bear\Desktop\liver segmentation\models\ortho_model"

io.logger_setup()
model = models.CellposeModel(gpu = True, pretrained_model = model_yx_path, pretrained_model_ortho = model_ortho_path)
image_cellpose = io.imread(file)
print(image_cellpose.shape)

channels = [2, 1]
masks, flows, styles = model.eval(image_cellpose, channels = channels, do_3D = True)
io.save_masks(image_cellpose, masks, flows, file, tif=True, channels = [2, 1], suffix='_cp_masks_trained_3d')
```
